```python
# Replacing the null values by the mean/mode in numeriacl/categorical column
data['total_tests'].fillna(4.005175e+06, inplace = True)
data['new_tests'].fillna(1.230359e+05, inplace = True)
data['total_tests_per_thousand'].fillna(1.971000e+00, inplace = True)
data['new_tests_per_thousand'].fillna(7.800000e-02, inplace = True)
data['new_tests_smoothed'].fillna(9.305900e+04, inplace = True)
data['new_tests_smoothed_per_thousand'].fillna(6.750000e-02, inplace = True)
data['stringency_index'].fillna(7.546000e+01, inplace = True)
data['tests_units'].fillna('samples tested', inplace = True)
```

```python
# No null values present
data.isnull().any()
```

```python
# Using datetime to convert dates into ordinal numbers
import datetime as dt
data["date"]=pd.to_datetime(data["date"])
data["date"]=data["date"].map(dt.datetime.toordinal)
```

```python
# Dropping categorical columns
data = data.drop(['iso_code', 'continent', 'location', 'tests_units'], axis = 1)
```

```python
# Importing packages for Modelling
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# Assigning the target variable
y = data['total_cases'].values

# Assigning the feature variable
X = data.drop('total_cases', axis = 1).values
```

```python
# # Using the train_test_split to split the data into 25% testsize and 75% trainsize

(X_train, X_test, y_train, y_test) = train_test_split(X, y , test_size = 0.25)

# Instantiating the Model
reg_all = LinearRegression()

# Fitting the traing data
reg_all.fit(X_train, y_train)

# Prdicting on the test data
y_pred1 = reg_all.predict(X_test)

# Printing the R^2 and RMSE values
print('R^2 = {}'.format(reg_all.score(X_test, y_test)))

# RMSE gives the accuracy of the model
print('rmse = {}'.format(mean_squared_error(y_pred1,y_test)))
```

```python
# Importing packages for modelling
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as MSE
```
